```python
# !pip install -U phidata groq gradio

import os
import gradio as gr
from phi.agent import Agent
from phi.model.groq import Groq

# 1. Configure API Key
os.environ["GROQ_API_KEY"] = "api-key"
# 2. Build the Agent
agent = Agent(
    model=Groq(id="llama-3.3-70b-versatile"),
    description="You are a Technical Assistant who explains AI concepts like Embeddings.",
    instructions=[
        "Always start with a high-level definition.",
        "Use a simple analogy to explain the concept.",
        "Conclude with a 'Pro-Tip' for developers.",
    ],
    markdown=True
)

# 3. Function that connects Agent → Gradio
def chat(user_message, history):
    response = agent.run(user_message, stream=False)
    return response.content

# 4. Build Gradio UI
demo = gr.ChatInterface(
    fn=chat,
    title="🤖 AI Concept Explainer",
    description="Ask me anything about AI — Embeddings, RAG, Transformers, Agents & more!",
    examples=[
        "Explain how embeddings work",
        "What is RAG in LLMs?",
        "How do AI Agents differ from chatbots?",
        "Explain Transformers with an analogy"
    ],
    theme=gr.themes.Soft()
)

# 5. Launch — share=True gives a public URL in Colab
demo.launch(share=True)
```
